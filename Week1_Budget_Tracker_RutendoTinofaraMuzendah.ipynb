```python
# AI-GENERATED BUDGET TRACKER CODE
# # Ask user for monthly income
income = float(input("Enter your monthly income: "))

# Ask user for up to 3 expenses
expense1 = float(input("Enter expense 1: "))
expense2 = float(input("Enter expense 2: "))
expense3 = float(input("Enter expense 3: "))

# Calculate total expenses
total_expenses = expense1 + expense2 + expense3

# Calculate remaining budget
remaining_budget = income - total_expenses

# Display results
print("\n--- Budget Summary ---")
print("Total Expenses:", total_expenses)
print("Remaining Budget:", remaining_budget)

# Show warning if budget is negative
if remaining_budget < 0:
    print("⚠️ Warning: Your budget is negative! You are overspending.")
else:
    print("✅ Good job! You are within your budget.")
```

## Part 4: Enhancement Challenge

Choose ONE feature to add using AI assistance:
- 🏷️ Expense categories (food, transport, entertainment)
- 📊 Show percentage of income spent
- ✅ Input validation (no negative numbers)
- 📝 Detailed budget summary with advice

```python
# ENHANCED BUDGET TRACKER
# # Ask the user for their monthly income
income = float(input("Enter your monthly income: "))

# Ask for up to 3 expenses
expense1 = float(input("Enter expense 1: "))
expense2 = float(input("Enter expense 2: "))
expense3 = float(input("Enter expense 3: "))

# Calculate total expenses
total_expenses = expense1 + expense2 + expense3

# Calculate remaining budget
remaining_budget = income - total_expenses

# Calculate percentage spent
percentage_spent = (total_expenses / income) * 100

# Display results
print("\n----- Budget Summary -----")
print("Total Expenses:", total_expenses)
print("Remaining Budget:", remaining_budget)
print("Percentage of Income Spent:", round(percentage_spent, 2), "%")

# Show warning if budget is negative
if remaining_budget < 0:
    print("⚠️ Warning: You are over your budget!")
else:
    print("✅ Good job! You are within your budget.")
```
